```python
def convert_to_float(df, column_name):

    df[column_name] = (
        df[column_name]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .str.replace(' ', '', regex=False)
    )
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce') 
    return df
```

```python
def replace_observations(df, column_name):

    replacements = {
        "Agente com FA não enviado": "Não enviado",
        "Agente com patrimônio líquido ajustado negativo": "Negativo"
    }
    df[column_name] = df[column_name].replace(replacements)
    return df

merged_df = replace_observations(merged_df, 'Observação')
```

```python
def clean_and_fill_columns(df, keep_columns, target_column, fill_column):

    df = df[keep_columns].copy()  # Cria uma cópia explícita para evitar o aviso

    df.loc[:, target_column] = df[target_column].fillna(df[fill_column])
    
    return df
```

```python
# Especificando as colunas a serem mantidas
columns_to_keep = ['Sigla', 'Razão Social', 'Classe', 'Fator de Alavancagem', 'Observação']

merged_df = clean_and_fill_columns(merged_df, columns_to_keep, 'Fator de Alavancagem', 'Observação')

merged_df = merged_df.drop(columns=['Observação'])
```

```python
comercializadores = semana[semana['Classe'] == 'Comercializador']
geradores = semana[semana['Classe'] == 'Gerador']
auto_produtores = semana[semana['Classe'] == 'Autoprodutor']
produtores_independentes = semana[semana['Classe'] == 'Produtor Independente']
consumidores_livres = semana[semana['Classe'] == 'Consumidor Livre']
consumidores_especiais = semana[semana['Classe'] == 'Consumidor Especial']
```

```python
def calculate_metrics(df, fator_column, observacao_column):

    total_records = len(df)
    nao_enviados = len(df[df[observacao_column] == "Não enviado"]) / total_records * 100
    nivel_participacao = 100 - nao_enviados  # Nível de Participação = 1 - % Não Enviados
    negativos = len(df[df[observacao_column] == "Negativo"]) / total_records * 100
    zeros = len(df[df[fator_column] == 0]) / total_records * 100
    greater_than_zero = len(df[df[fator_column] > 0]) / total_records * 100
    pla_zero = len(
    df[
        (df[fator_column].isna()) &
        (df[observacao_column].str.strip() == "")
    ]
) / total_records * 100
    regularizado = len(
        df[df[observacao_column] == "Agente com FA enviado na regularização"]
    ) / total_records * 100
    mean_fator = df[fator_column].mean()
    mean_positive_fator = df[df[fator_column] > 0][fator_column].mean()
    std_dev = df[fator_column].std()

    total_percentage = (
        nao_enviados
        + negativos
        + zeros
        + greater_than_zero
        + pla_zero
        + regularizado
        )
    
    return {
        "Total Registros": total_records,
        "% Não Enviados": nao_enviados,
        "% Participação": nivel_participacao,
        "% Negativos": negativos,
        "% Zeros": zeros,
        "% Maior que Zero": greater_than_zero,
        "% PLA zero": pla_zero,
        "% Regularizado": regularizado,
        "Soma Total %": total_percentage,
        "Média Fator de Alavancagem": mean_fator,
        "Média Fator > 0": mean_positive_fator,
        "Desvio Padrão": std_dev,
    }
```

```python
# Calculando as métricas para cada classe
metrics_comercializadores = calculate_metrics(comercializadores, "Fator de Alavancagem", "Observação")
metrics_geradores = calculate_metrics(geradores, "Fator de Alavancagem", "Observação")
metrics_auto_produtores = calculate_metrics(auto_produtores, "Fator de Alavancagem", "Observação")
metrics_produtores_independentes = calculate_metrics(produtores_independentes, "Fator de Alavancagem", "Observação")
metrics_consumidores_livres = calculate_metrics(consumidores_livres, "Fator de Alavancagem", "Observação")
metrics_consumidores_especiais = calculate_metrics(consumidores_especiais, "Fator de Alavancagem", "Observação")
```

```python
# Exibindo as métricas calculadas para cada classe
print("Métricas - Comercializadores:")
for key, value in metrics_comercializadores.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")
    
print("\nMétricas - Geradores:")
for key, value in metrics_geradores.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")

print("\nMétricas - Auto Produtores:")
for key, value in metrics_auto_produtores.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")

print("\nMétricas - Produtores Independentes:")
for key, value in metrics_produtores_independentes.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")

print("\nMétricas - Consumidores Livres:")
for key, value in metrics_consumidores_livres.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")

print("\nMétricas - Consumidores Especiais:")
for key, value in metrics_consumidores_especiais.items():
    print(f"{key}: {value:.4f}" if isinstance(value, (float, int)) else f"{key}: {value}")
```

```python
# Criando um dicionário com as métricas para cada classe
metrics_data = {
    "Comercializadores": metrics_comercializadores,
    "Geradores": metrics_geradores,
    "Auto Produtores": metrics_auto_produtores,
    "Produtores Independentes": metrics_produtores_independentes,
    "Consumidores Livres": metrics_consumidores_livres,
    "Consumidores Especiais": metrics_consumidores_especiais
}

# Convertendo o dicionário em um DataFrame
metrics_df = pd.DataFrame(metrics_data)

metrics_df = metrics_df.reindex([
    "Total Registros",
    "% Não Enviados",
    "% Participação",
    "% Negativos",
    "% Zeros",
    "% Maior que Zero",
    "% PLA zero",
    "% Regularizado",
    "Soma Total %",
    "Média Fator de Alavancagem",
    "Média Fator > 0",
    "Desvio Padrão"
])

# Dicionário definindo o número de casas decimais para cada linha
decimal_places_per_row = {
    "Total Registros": 0,
    "% Não Enviados": 4,
    "% Participação": 4,
    "% Negativos": 4,
    "% Zeros": 4,
    "% Maior que Zero": 4,
    "% PLA zero": 4,
    "% Regularizado": 4,
    "Soma Total %": 4,
    "Média Fator de Alavancagem": 2,
    "Média Fator > 0": 2,
    "Desvio Padrão": 2,
}

# Aplicando diferentes arredondamentos para cada linha e armazenando como strings
for row, decimals in decimal_places_per_row.items():
    metrics_df.loc[row] = metrics_df.loc[row].apply(
        lambda x: f"{round(x, decimals):.{decimals}f}" if isinstance(x, (float, int)) else x
    )
```

```python
metrics_df.index.name = "Métrica"
metrics_df = metrics_df.reset_index()
```
